## Introduction

We will be investigating the relation between mpg and displacement in cars from this dataset.

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
cars = pd.read_csv('../input/cars-csv/cars.csv')
cars.head()
````

## Glossary
- **mpg**: stands for miles per gallon, a higher mpg means a higher fuel efficiency.
- **cylinders**: where fuel is combusted and power generated in an engine.
- **displacement**: means an engine can move more air and fuel, giving it the potential to make more power.
- **horsepower**: the power an engine produces.
- **weight**: weight of the car in pounds.
- **acceleration**: time to accelerate from 0 to 60 mph in seconds.
- **model_year**: the year the car was manufactured.
- **origin**: the country the car was manufactured
- **name**: car name

````python
cars.describe()
````

```horsepower``` contains missing values, but since we are not working with ```horsepower```, we will not fill the missing data.

## Hypothesis


````python
cars = cars[['mpg', 'displacement']]
cars.head()
````

````python
plt.scatter(x=cars['mpg'], y=cars['displacement'])
plt.show()
````

## Summary
It is sufficient to conclude that cars with a high engine displacement have a low mpg and vice versa, meaning they consume more fuel. There are also some outliers which could be investigated.
